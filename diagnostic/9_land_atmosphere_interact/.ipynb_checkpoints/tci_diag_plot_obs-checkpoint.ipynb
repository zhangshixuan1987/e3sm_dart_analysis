```python
import os
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from scipy.stats import ttest_ind
from matplotlib import cm
from matplotlib import ticker
from scipy.ndimage import gaussian_filter

try:
    import cmocean
    cmap_default = cmocean.cm.balance
except ImportError:
    cmap_default = 'RdBu_r'

class TCIPlotter:
    def __init__(self, tci_files_dict, landmask_file=None, compute_ensemble_mean=True):
        """
        tci_files_dict: dict of {experiment_name: path_to_file or list of files}
        landmask_file: optional path to landmask NetCDF file
        compute_ensemble_mean: whether to average over ensemble members
        """
        self.tci_files = tci_files_dict
        self.compute_ensemble_mean = compute_ensemble_mean
        self.landmask = self._load_landmask(landmask_file)
        self.tci_data = self._load_tci_data()

    def _load_landmask(self, landmask_file):
        if not landmask_file or not os.path.exists(landmask_file):
            return None
        ds = xr.open_dataset(landmask_file)
        landmask = ds[list(ds.data_vars)[0]]
        if 'lon' in landmask.coords and landmask.lon.max() > 180:
            landmask['lon'] = ((landmask['lon'] + 180) % 360) - 180
            landmask = landmask.sortby('lon')
        return landmask

    def _load_tci_data(self):
        tci_data = {}
        for exp, files in self.tci_files.items():
            files = [files] if isinstance(files, str) else files
            da_list = [xr.open_dataset(f)['TCI'] for f in files]
            da_ens = xr.concat(da_list, dim='time') if len(da_list) > 1 else da_list[0].expand_dims('time') if 'time' not in da_list[0].dims else da_list[0]

            if self.landmask is not None:
                mask_interp = self.landmask.interp_like(da_ens, method='nearest')
                da_ens = da_ens.where(mask_interp > 0.5)

            if self.compute_ensemble_mean:
                da_ens = da_ens.mean(dim='time', skipna=True)

            tci_data[exp] = da_ens
        return tci_data

    def plot_spatial_mean(self, region=None, savepath=None):
        fig, ax = plt.subplots()
        for exp, da in self.tci_data.items():
            if region:
                da = da.sel(lat=slice(*region['lat']), lon=slice(*region['lon']))
            mean_tci = da.mean(dim=['lat', 'lon'], skipna=True)
            ax.bar(exp, mean_tci.values)
        ax.set_title("Mean TCI by Experiment")
        ax.set_ylabel("TCI")
        if savepath:
            plt.savefig(savepath, bbox_inches='tight')
        plt.show()

    def plot_tci_map(self, tag=None, compare=False, significance_level=0.05,
                     savepath=None, use_robinson=True, smooth=False):
        exps = list(self.tci_data.keys())
        if not exps:
            print("[ERROR] No data to plot.")
            return
        if compare and len(exps) != 2:
            raise ValueError("Comparison requires exactly 2 experiments.")

        proj = ccrs.Robinson() if use_robinson else ccrs.PlateCarree()
        fig, ax = plt.subplots(figsize=(10, 5), subplot_kw={'projection': proj})

        da1 = self.tci_data[exps[0]]
        if compare:
            da2 = self.tci_data[exps[1]]
            diff = da1 - da2
            data = gaussian_filter(diff.values, sigma=1.0) if smooth else diff.values
            lon, lat = np.meshgrid(diff['lon'], diff['lat'])
            im = ax.contourf(lon, lat, data, transform=ccrs.PlateCarree(), cmap=cmap_default, levels=21)

            ax.set_title(f"{exps[0]} - {exps[1]}")
            if not self.compute_ensemble_mean:
                t1 = xr.concat([xr.open_dataset(f)['TCI'] for f in self.tci_files[exps[0]]], dim='time')
                t2 = xr.concat([xr.open_dataset(f)['TCI'] for f in self.tci_files[exps[1]]], dim='time')
                pval = xr.apply_ufunc(
                    lambda x, y: ttest_ind(x, y, axis=0, nan_policy='omit').pvalue,
                    t1, t2, input_core_dims=[['time'], ['time']], vectorize=True
                )
                sig_mask = pval < significance_level
                ax.contourf(lon, lat, sig_mask, levels=[0.5, 1.5], hatches=['...', None],
                            colors='none', transform=ccrs.PlateCarree())

        else:
            data = gaussian_filter(da1.values, sigma=1.0) if smooth else da1.values
            lon, lat = np.meshgrid(da1['lon'], da1['lat'])
            im = ax.contourf(lon, lat, data, transform=ccrs.PlateCarree(), cmap=cmap_default, levels=21)
            ax.set_title(f"{exps[0]} (mean)" if self.compute_ensemble_mean else exps[0])

        ax.coastlines()
        ax.add_feature(cfeature.BORDERS, linewidth=0.5)
        cbar = fig.colorbar(im, ax=ax, orientation='horizontal', pad=0.05, shrink=0.8)
        cbar.set_label('TCI')
        if savepath:
            plt.savefig(savepath, bbox_inches='tight', dpi=300)
        plt.show()
```
