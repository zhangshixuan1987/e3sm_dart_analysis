```python
class ShockDecayAnalyzer:
    def __init__(self, data_dir, var_list, ensemble_members, filename_template,
                 ref_period, shock_period, vertical_dim='plev'):
        self.data_dir = data_dir
        self.var_list = var_list
        self.ensemble_members = ensemble_members
        self.filename_template = filename_template
        self.ref_period = ref_period
        self.shock_period = shock_period
        self.vertical_dim = vertical_dim
        self.results = {}

    def exp_decay(self, t, a0, tau, c):
        return a0 * np.exp(-t / tau) + c

    def _load_variable(self, varname, ens):
        fpath = os.path.join(self.data_dir, self.filename_template.format(ens=ens))
        ds = xr.open_dataset(fpath)
        if varname not in ds:
            raise KeyError(f"{varname} not found in {fpath}")
        var = ds[varname]
        if "PRECT" in varname and "units" in var.attrs and "kg" in var.attrs["units"]:
            var = var * 86400  # Convert to mm/day
        return var

    def _global_weighted_mean(self, da):
        weights = np.cos(np.deg2rad(da.lat))
        return da.weighted(weights).mean(dim=("lat", "lon"))

    def _compute_anomaly(self, da_mean):
        clim = da_mean.sel(time=self.ref_period).groupby("time.dayofyear").mean("time")
        shock = da_mean.sel(time=self.shock_period)
        return shock.groupby("time.dayofyear") - clim

    def _fit_decay(self, t, y):
        mask = ~np.isnan(y)
        if np.sum(mask) < 10:
            return np.nan
        try:
            popt, _ = curve_fit(self.exp_decay, t[mask], y[mask], p0=(y[mask][0], 5, 0))
            return popt[1], popt  # tau
        except RuntimeError:
            return np.nan, None

    def analyze_variable(self, varname, vertical_profile=False):
        self.results[varname] = {}

        for ens in self.ensemble_members:
            print(f"Analyzing {varname} - {ens}")
            try:
                var = self._load_variable(varname, ens)
            except KeyError as e:
                print(f"  Skipping: {e}")
                continue

            if vertical_profile and self.vertical_dim in var.dims:
                # For each level
                taus = {}
                for lev in var[self.vertical_dim]:
                    da_level = var.sel({self.vertical_dim: lev})
                    mean = self._global_weighted_mean(da_level)
                    anom = self._compute_anomaly(mean).rolling(time=5, center=True).mean()
                    t = np.arange(anom.time.size)
                    tau, popt = self._fit_decay(t, anom.values)
                    taus[float(lev.values)] = tau

                    # Optional plot
                    self._plot_fit(t, anom.values, popt, varname, ens, lev=float(lev.values))

                self.results[varname][ens] = taus
            else:
                mean = self._global_weighted_mean(var)
                anom = self._compute_anomaly(mean).rolling(time=5, center=True).mean()
                t = np.arange(anom.time.size)
                tau, popt = self._fit_decay(t, anom.values)
                self.results[varname][ens] = tau

                # Optional plot
                self._plot_fit(t, anom.values, popt, varname, ens)

    def _plot_fit(self, t, y, popt, varname, ens, lev=None):
        plt.figure(figsize=(8, 4))
        plt.plot(t, y, label="Anomaly", color="blue")
        if popt is not None:
            plt.plot(t, self.exp_decay(t, *popt), "r--", label=f"Fit (τ ≈ {popt[1]:.1f} d)")
        label = f"{varname}_{ens}" if lev is None else f"{varname}_{ens}_lev{lev:.0f}"
        plt.title(f"Decay of {label}")
        plt.xlabel("Days since shock")
        plt.ylabel("Anomaly")
        plt.axhline(0, color="gray", linestyle=":")
        plt.legend()
        plt.tight_layout()
        plt.savefig(f"shock_decay_{label}.png")
        plt.close()

    def summarize_results(self):
        summary = pd.DataFrame(self.results).T
        return summary
```
