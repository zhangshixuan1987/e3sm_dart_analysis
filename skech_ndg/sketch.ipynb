```python
import matplotlib.pyplot as plt
import numpy as np

def draw_column(ax, x, label, mode="default"):
    # Column box
    ax.add_patch(plt.Rectangle((x,0),1,6, fill=False, linewidth=2))
    
    # PBL shading
    ax.add_patch(plt.Rectangle((x,0),1,1.5, color='lightgray', alpha=0.4))
    
    ax.text(x+0.5,6.4,label,ha='center')
    
    if mode == "default":
        for y in [1,2.5,4.5]:
            ax.arrow(x+0.2,y,0.5,0,head_width=0.2,color='blue')
    
    elif mode == "revised":
        ax.arrow(x+0.2,1,0.3,0,head_width=0.2,color='blue')  # weak PBL
        ax.arrow(x+0.2,3,0.7,0,head_width=0.2,color='blue')  # strong mid
        ax.plot([x+0.3,x+0.8],[5,5.5],color='red',linewidth=2)

    elif mode == "improved":
        ax.text(x+0.5,3,"☁",fontsize=40,ha='center')
        ax.text(x+0.5,1,"☂",fontsize=25,ha='center')

fig, ax = plt.subplots(figsize=(10,5))

draw_column(ax,0,"Default (Eq.3)","default")
draw_column(ax,2,"Revised (Eq.4)","revised")
draw_column(ax,4,"Improved Skill","improved")

# Weighting curve
z = np.linspace(0,6,200)
weight = 0.5*(1+np.tanh((z-2.5)))
ax.plot(6+weight,z,color='blue')
ax.text(6.5,5.5,"Weight")

ax.set_xlim(-0.5,8)
ax.set_ylim(0,7)
ax.axis('off')
plt.show()
```
