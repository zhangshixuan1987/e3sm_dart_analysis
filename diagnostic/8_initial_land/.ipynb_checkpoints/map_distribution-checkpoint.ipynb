```python
data_dict = {
    'CPC_SOM': {
            'path': '/pscratch/sd/z/zhan391/e3sm_dart/Observations/CPC_SOM/monthly',
            'template': 'SOILWATER_10CM.daily.%(year).nc',
            'frequency': 'monthly',
            'nens': 1
    },
    'ESA_CCI': {
            'path': '/pscratch/sd/z/zhan391/e3sm_dart/Observations/ESA_CCI/daily',
            'template': 'H2OSOI.daily.%(year).nc',
            'frequency': 'daily',
            'nens': 1
    },
    'CTRLEN10': {
            'path': '/pscratch/sd/z/zhan391/e3sm_dart/CTRLEN10_15day_F20TR_ne30pg2_r05_IcoswISC30E3r5_compy/archive/post/lnd/180x360_aave/clim',
            'template': 'H2OSOI.%(ensemble)s.%(year).nc',
            'frequency': 'daily',
            'nens': 10
    },
    'CAPTEN10': {
            'path': '/pscratch/sd/z/zhan391/e3sm_dart/CTRLEN10_15day_F20TR_ne30pg2_r05_IcoswISC30E3r5_compy/archive/post/lnd/180x360_aave/clim',
            'template': 'H2OSOI.%(ensemble)s.%(year).nc',
            'frequency': 'daily',
            'nens': 10
    },
    'DARTEN20': {
            'path': '/pscratch/sd/z/zhan391/e3sm_dart/DARTEN20_15day_F20TR_ne30pg2_r05_IcoswISC30E3r5_compy/archive/post/lnd/180x360_aave/clim',
            'template': 'H2OSOI.%(ensemble)s.%(year).nc',
            'frequency': 'daily',
            'nens': 20
    },
    'DARTEN40': {
            'path': '/pscratch/sd/z/zhan391/e3sm_dart/DARTEN40_15day_F20TR_ne30pg2_r05_IcoswISC30E3r5_compy/archive/post/lnd/180x360_aave/clim',
            'template': 'H2OSOI.%(ensemble)s.%(year).nc',
            'frequency': 'daily',
            'nens': 40
    }
}
```

```python
class SoilMoistureDifferencePlotter:
    def __init__(self, data_dict, landmask_file=None, soilayer_file=None,
                 variable='H2OSOI', dz_var='DZSOI', landmask_var='landfrac',
                 depth_cm=10, mask_land=True, confidence_level = 0.05,
                 use_mass_units=True):
        self.data_dict = data_dict
        self.variable = variable
        self.dz_var = dz_var
        self.landmask_var = landmask_var
        self.depth_cm = depth_cm
        self.mask_land = mask_land 
        self.significance_threshold = 0  #
        self.confidence_level = confidence_level 
        self.use_mass_units = use_mass_units
        self.water_density = 1000.0  # kg/m³
        
        if landmask_file and os.path.exists(landmask_file):
            landmask = xr.open_dataset(landmask_file)[landmask_var]
            if landmask.max() > 1.1:
                landmask = landmask / 100.0
            self.landmask = landmask.where(landmask > 0.1) if mask_land else landmask.where(landmask <= 0.5)
            
        self.fixed_dz = None
        if soilayer_file and os.path.exists(soilayer_file):
            self.fixed_dz = xr.open_dataset(soilayer_file)[dz_var]

    def _load_dataset(self, info, year, ensemble=None):
        template = info['template']
        path = info['path']
        if ensemble:
            filename = template % {'year': year, 'ensemble': ensemble}
        else:
            filename = template % {'year': year}
        full_path = os.path.join(path, filename)
        if not os.path.exists(full_path):
            raise FileNotFoundError(f"[ERROR] File not found: {full_path}")
        return xr.open_dataset(full_path)

    def _integrate_top_layers(self, ds):
        sm = ds[self.variable]
    
        if self.dz_var in ds:
            dz = ds[self.dz_var]
            if "time" in dz.dims:
                dz = dz.isel(time=0)
        elif self.fixed_dz is not None:
            dz = self.fixed_dz
        else:
            raise ValueError("No valid DZSOI found in dataset or external source.")

        dz_cumsum = dz.cumsum(dim='levgrnd')
        top_mask = dz_cumsum <= (self.depth_cm / 100.0)
        sm_top = sm.where(top_mask)
        dz_masked = dz.where(top_mask)
    
        if self.use_mass_units:
            # Convert to kg/m²: SM * DZ * ρ_water
            sm_mass = sm_top * dz_masked * self.water_density
            integrated = sm_mass.sum(dim='levgrnd')
            integrated.attrs['units'] = 'kg/m2'
        else:
            # Keep volumetric average (m³/m³)
            integrated = (sm_top * dz_masked).sum(dim='levgrnd') / dz_masked.sum(dim='levgrnd')
            integrated.attrs['units'] = 'm3/m3'
    
        if self.landmask is not None:
            integrated = integrated.where(self.landmask > 0.1)
    
        return integrated

    def _load_model_ensemble_snapshot(self, model_info, target_date):
        nens = model_info['nens']
        ens_data = []
        for n in range(1, nens + 1):
            ens = f'EN{n:02d}'
            ds = self._load_dataset(model_info, target_date.year, ensemble=ens)
            ds_day = ds.sel(time=target_date)
            sm_top = self._integrate_top_layers(ds_day)
            ens_data.append(sm_top.squeeze())
        return xr.concat(ens_data, dim='ensemble')

    def _load_obs_snapshot(self, obs_info, target_date):
        ds = self._load_dataset(obs_info, target_date.year)
        obs = ds[self.variable].sel(time=target_date)
        if self.use_mass_units:
            obs = obs * 1000.0 * self.depth_cm / 100.0 
        if self.landmask is not None:
            obs = obs.where(self.landmask > 0.1)
        return obs
        
    def _apply_mask(self, da):
        return da.where(self.landmask > 0.5) if self.landmask is not None else da
    
    def _compute_significance_mask(self, ensemble_data, obs_mean, alpha=0.05, threshold=None):
        """
        Compute significance mask (True = significant) and mask where soil moisture is too small.
        """
        if threshold is None:
            threshold = self.significance_threshold
    
        diff = ensemble_data - obs_mean
        tstat, pvals = ttest_1samp(diff.transpose("ensemble", "lat", "lon").values,
                                   popmean=0.0, axis=0, nan_policy='omit')
    
        sig_mask = xr.DataArray(pvals < alpha, coords=obs_mean.coords, dims=obs_mean.dims)
    
        model_mean = ensemble_data.mean(dim='ensemble')
        valid_mask = (abs(model_mean) > threshold) & (abs(obs_mean) > threshold)
    
        sig_mask = sig_mask.where(valid_mask)
    
        return sig_mask
      
    def plot_bias_and_stddev_multi_panel(self, model_keys, target_date='2012-01-01',
                                         obs_key='ESA_CCI',
                                         bias_levels=None, spread_levels=None,
                                         cmap_bias='RdBu_r', cmap_std='viridis',
                                         figsize=(12, 5), savepath=None):
        """
        Plot mean bias (model - obs) and ensemble spread (std dev) in two rows.
        """
        target_date = pd.to_datetime(target_date)
        ncols = len(model_keys)
    
        # Default discrete levels
        if bias_levels is None:
            bias_levels = [-0.5, -0.3, -0.1, -0.05, -0.01, 0.01, 0.05, 0.1, 0.3, 0.5]
        if spread_levels is None:
            spread_levels = [0.005, 0.01, 0.02, 0.04, 0.06, 0.08, 0.1]
            
        # Discrete colormaps
        bias_cmap = mpl.colors.ListedColormap(plt.get_cmap(cmap_bias)(np.linspace(0, 1, len(bias_levels) - 1)))
        bias_norm = mcolors.BoundaryNorm(bias_levels, ncolors=bias_cmap.N)
    
        spread_cmap = mpl.colors.ListedColormap(plt.get_cmap(cmap_std)(np.linspace(0, 1, len(spread_levels) - 1)))
        spread_norm = mcolors.BoundaryNorm(spread_levels, ncolors=spread_cmap.N)

        fig, axes = plt.subplots(
            nrows=2, ncols=ncols, figsize=figsize,
            subplot_kw={'projection': ccrs.PlateCarree()},
            constrained_layout=False
        )
    
        for j, model_key in enumerate(model_keys):
            model_info = self.data_dict[model_key]
            obs_info = self.data_dict[obs_key]
    
            ens_stack = self._load_model_ensemble_snapshot(model_info, target_date)
            obs_field = self._load_obs_snapshot(obs_info, target_date)
    
            if ens_stack.shape[1:] != obs_field.shape:
                ens_stack = ens_stack.interp_like(obs_field)
    
            model_mean = ens_stack.mean(dim='ensemble')
            model_std = ens_stack.std(dim='ensemble')
            bias = model_mean - obs_field
            
            valid_mask = np.isfinite(bias.values)
            bias_mean_val = float(bias.values[valid_mask].mean())
            rmse_val = float(np.sqrt((bias.values[valid_mask] ** 2).mean()))
            
            # --- Row 0: Bias ---
            ax0 = axes[0, j]
            bias.plot.pcolormesh(
                ax=ax0, transform=ccrs.PlateCarree(),
                cmap=bias_cmap, norm=bias_norm, add_colorbar=False
            )
            ax0.set_title(f"{model_key} - {obs_key} (Bias)")
            ax0.coastlines()
            ax0.set_xticks(np.arange(-180, 181, 60), crs=ccrs.PlateCarree())
            ax0.set_yticks(np.arange(-90, 91, 30), crs=ccrs.PlateCarree())
            ax0.set_xticklabels([])
            ax0.tick_params(labelsize=8)
            if j == 0:
                ax0.set_ylabel("Latitude")
            else:
                ax0.set_ylabel("")
                ax0.set_yticklabels([])
    
            # Bias stats
            stats_text = f"Mean: {bias_mean_val:.3f}\nRMSE: {rmse_val:.3f}"
            ax0.text(
                0.98, 0.02, stats_text,
                transform=ax0.transAxes, fontsize=8,
                va='bottom', ha='right',
                bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3')
            )
    
            # Significance mask
            sig_mask = self._compute_significance_mask(ens_stack, obs_field, alpha=self.confidence_level)
            sig_mask.plot.contourf(
                ax=ax0, transform=ccrs.PlateCarree(),
                colors='none', hatches=['...', None],
                add_colorbar=False, linewidths=0
            )
            if j == 0:
                ax0.set_ylabel("Latitude")
            else:
                ax0.set_ylabel("")
                ax0.set_yticklabels([])
                
            # --- Row 1: Spread ---
            ax1 = axes[1, j]
            model_std.plot.pcolormesh(
                ax=ax1, transform=ccrs.PlateCarree(),
                cmap=spread_cmap, norm=spread_norm, add_colorbar=False
            )
            ax1.set_title(f"{model_key} (Spread)")
            ax1.coastlines()
            ax1.set_xticks(np.arange(-180, 181, 60), crs=ccrs.PlateCarree())
            ax1.set_yticks(np.arange(-90, 91, 30), crs=ccrs.PlateCarree())
            ax1.set_xlabel("Longitude")
            ax1.tick_params(labelsize=8)
            if j == 0:
                ax1.set_ylabel("Latitude")
            else:
                ax1.set_ylabel("")
                ax1.set_yticklabels([])

        # Adjust plot layout to leave space on right
        plt.subplots_adjust(wspace=0.1, right=0.88, hspace=0.2)
        
        # Add vertical colorbar for bias
        cbar_ax1 = fig.add_axes([0.90, 0.55, 0.012, 0.22])  # [left, bottom, width, height]
        sm1 = mpl.cm.ScalarMappable(norm=bias_norm, cmap=bias_cmap)
        sm1.set_array(np.array([0]))
        fig.colorbar(
            sm1, cax=cbar_ax1,
            ticks=bias_levels,
            label="Bias (kg m$^{-2}$)"
        )
        
        # Add vertical colorbar for spread
        cbar_ax2 = fig.add_axes([0.90, 0.12, 0.012, 0.22])
        sm2 = mpl.cm.ScalarMappable(norm=spread_norm, cmap=spread_cmap)
        sm2.set_array(np.array([0]))
        fig.colorbar(
            sm2, cax=cbar_ax2,
            ticks=spread_levels,
            label="Spread (kg m$^{-2}$)"
        )
        
        #plt.tight_layout(rect=[0, 0.03, 1, 1])
        plt.subplots_adjust(hspace=0.15, wspace=0.05)
        plt.show()
    
        if savepath:
            fig.savefig(savepath, dpi=600, bbox_inches='tight')
            print(f"[SAVED] Figure saved to: {savepath}")
```

```python
import cartopy.feature as cfeature

def plot_bias_and_stddev_multi_panel(self, model_keys, target_date='2012-01-01',
                                     obs_key='ESA_CCI',
                                     bias_levels=None, spread_levels=None,
                                     cmap_bias='RdBu_r', cmap_std='viridis',
                                     figsize=(12, 5), savepath=None):
    """
    Plot mean bias (model - obs) and ensemble spread (std dev) in two rows.
    Uses Robinson projection for improved land visibility.
    """
    target_date = pd.to_datetime(target_date)
    ncols = len(model_keys)

    if bias_levels is None:
        bias_levels = [-0.5, -0.3, -0.1, -0.05, -0.01, 0.01, 0.05, 0.1, 0.3, 0.5]
    if spread_levels is None:
        spread_levels = [0.005, 0.01, 0.02, 0.04, 0.06, 0.08, 0.1]

    # Discrete colormaps
    bias_cmap = mpl.colors.ListedColormap(plt.get_cmap(cmap_bias)(np.linspace(0, 1, len(bias_levels) - 1)))
    bias_norm = mcolors.BoundaryNorm(bias_levels, ncolors=bias_cmap.N)

    spread_cmap = mpl.colors.ListedColormap(plt.get_cmap(cmap_std)(np.linspace(0, 1, len(spread_levels) - 1)))
    spread_norm = mcolors.BoundaryNorm(spread_levels, ncolors=spread_cmap.N)

    projection = ccrs.Robinson()

    fig, axes = plt.subplots(
        nrows=2, ncols=ncols, figsize=figsize,
        subplot_kw={'projection': projection},
        constrained_layout=False
    )

    for j, model_key in enumerate(model_keys):
        model_info = self.data_dict[model_key]
        obs_info = self.data_dict[obs_key]

        ens_stack = self._load_model_ensemble_snapshot(model_info, target_date)
        obs_field = self._load_obs_snapshot(obs_info, target_date)

        if ens_stack.shape[1:] != obs_field.shape:
            ens_stack = ens_stack.interp_like(obs_field)

        model_mean = ens_stack.mean(dim='ensemble')
        model_std = ens_stack.std(dim='ensemble')
        bias = model_mean - obs_field

        valid_mask = np.isfinite(bias.values)
        bias_mean_val = float(bias.values[valid_mask].mean())
        rmse_val = float(np.sqrt((bias.values[valid_mask] ** 2).mean()))

        # --- Row 0: Bias ---
        ax0 = axes[0, j]
        bias.plot.pcolormesh(
            ax=ax0, transform=ccrs.PlateCarree(),
            cmap=bias_cmap, norm=bias_norm, add_colorbar=False
        )
        ax0.set_title(f"{model_key} - {obs_key} (Bias)")
        ax0.coastlines()
        ax0.add_feature(cfeature.LAND, facecolor='lightgray', zorder=0)
        ax0.gridlines(draw_labels=False, linewidth=0.3, color='gray', alpha=0.5)
        ax0.tick_params(labelsize=8)
        if j == 0:
            ax0.set_ylabel("Latitude")

        stats_text = f"Mean: {bias_mean_val:.3f}\nRMSE: {rmse_val:.3f}"
        ax0.text(
            0.98, 0.02, stats_text,
            transform=ax0.transAxes, fontsize=8,
            va='bottom', ha='right',
            bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3')
        )

        sig_mask = self._compute_significance_mask(ens_stack, obs_field, alpha=self.confidence_level)
        sig_mask.plot.contourf(
            ax=ax0, transform=ccrs.PlateCarree(),
            colors='none', hatches=['...', None],
            add_colorbar=False, linewidths=0
        )

        # --- Row 1: Spread ---
        ax1 = axes[1, j]
        model_std.plot.pcolormesh(
            ax=ax1, transform=ccrs.PlateCarree(),
            cmap=spread_cmap, norm=spread_norm, add_colorbar=False
        )
        ax1.set_title(f"{model_key} (Spread)")
        ax1.coastlines()
        ax1.add_feature(cfeature.LAND, facecolor='lightgray', zorder=0)
        ax1.gridlines(draw_labels=False, linewidth=0.3, color='gray', alpha=0.5)
        ax1.set_xlabel("Longitude")
        ax1.tick_params(labelsize=8)
        if j == 0:
            ax1.set_ylabel("Latitude")

    # Adjust layout to make room for external colorbars
    plt.subplots_adjust(wspace=0.1, right=0.88, hspace=0.2)

    # Add vertical colorbar for bias
    cbar_ax1 = fig.add_axes([0.90, 0.55, 0.015, 0.35])
    sm1 = mpl.cm.ScalarMappable(norm=bias_norm, cmap=bias_cmap)
    sm1.set_array(np.array([0]))
    fig.colorbar(sm1, cax=cbar_ax1, ticks=bias_levels, label="Bias (kg m$^{-2}$)")

    # Add vertical colorbar for spread
    cbar_ax2 = fig.add_axes([0.90, 0.12, 0.015, 0.35])
    sm2 = mpl.cm.ScalarMappable(norm=spread_norm, cmap=spread_cmap)
    sm2.set_array(np.array([0]))
    fig.colorbar(sm2, cax=cbar_ax2, ticks=spread_levels, label="Spread (kg m$^{-2}$)")

    plt.show()

    if savepath:
        fig.savefig(savepath, dpi=600, bbox_inches='tight')
        print(f"[SAVED] Figure saved to: {savepath}")
```
