```python
def read_model_var(var,frequency,exp,exp_dict,path):
    rpath = glob.glob(os.path.join(
        path,exp_dict[exp]['run'],exp_dict[exp]['clim'],
        '{}.*.{}.nc'.format(var,exp_dict[exp]['period']))
    )
    dr = xc.open_mfdataset(rpath, combine='nested', concat_dim='time')
    rlats,rlons = dr['lat'].values, dr['lon'].values
    if np.min(rlons) > -1.0:
        dr= dr.assign_coords(lon=((rlons + 180) % 360 - 180) )
        dr = dr.sortby(dr.lon)
        rlons = dr['lon'].values
    if var == 'PRECT':
        data_dict = dr[var][:,:,:] * 24.0 * 3600.0 * 1000.0 
    elif var in ['TREFHT','TS','T850']:
        data_dict = dr[var][:,:,:] - 273.15 
    elif var in ['PSL','PS']:
        data_dict = np.mean(refds[var],axis=0) / 100.0
    else:
        data_dict = dr[var][:,:,:]
    data_dict = data_dict.chunk({'time':-1,'lat': -1, 'lon': -1})

    return data_dict 
```

```python
def define_region(regnam='global'):
    reg_dict = {'global':[(-90,90),(-180,180)],
                'Atlantic':[(10,90),(-100,10)],
                'CONUS':[(25,50),(235,295)],
                'Antarctic':[(-90,-50),(-180,180)],
                'PolarN':[(50,90),(-180,180)],
                #'Greenland':[(59,83),(-74,-11)],
                'Greenland':[(60,85),(-75,-10)],
               }
    return reg_dict[regnam]
```

```python
def generate_coslat_weight(ds):
    # make weights as cosine of the latitude and broadcast
    weights = np.cos(np.deg2rad(ds.lat))
    _, weights = xr.broadcast(ds,weights)
    return weights
```

```python
def derive_hor_metrics_data(var,dso1,dsm1):
    mec_dict = dict()
    if len(dso1) > 0:
        dsm_mean = np.mean(dsm1, axis=0)
        dsm_std = np.std(dsm1, axis=0)
        dsm = dsm_mean.to_dataset(name=var).bounds.add_missing_bounds(axes=("X", "Y"))
        print('mod_min_max_{}_{}'.format(dsm[var].min().values,dsm[var].max().values))
        weights = generate_coslat_weight(dsm[var])
        mec_dict['mean'] = np.sqrt(dsm.spatial.average(var,weights="generate")[var])  
        dsm[var] = dsm[var]**2
        mec_dict['rmsm'] = np.sqrt(dsm.spatial.average(var,weights="generate")[var])
                              
        dso = dso1.to_dataset(name=var).bounds.add_missing_bounds(axes=("X", "Y"))
        print('obs_min_max_{}_{}'.format(dso[var].min().values,dso[var].max().values))
        dso[var] = dso[var]**2 
        mec_dict['rmso'] = np.sqrt(dso.spatial.average(var, weights="generate")[var])
        mec_dict['rmse'] = xs.rmse(dso1,dsm_mean,dim=["lat", "lon"],weights=weights.astype("float64"),skipna=True)
        mec_dict['pcor'] = xs.pearson_r(dso1,dsm_mean,dim=["lat", "lon"],weights=weights.astype("float64"),skipna=True)
        dat_dict = dsm1.copy() - dso1
        dat_mean = np.mean(dat_dict, axis=0)
        dat_std = np.std(dat_dict, axis=0)
    else:
        dsm_mean = np.mean(dsm1, axis=0)
        dsm_std = np.std(dsm1, axis=0)
        dsm = dsm_mean.to_dataset(name=var).bounds.add_missing_bounds(axes=("X", "Y"))
        print('mod_min_max_{}_{}'.format(dsm[var].min().values,dsm[var].max().values))
        weights = generate_coslat_weight(dsm[var])
        mec_dict['mean'] = np.sqrt(dsm.spatial.average(var,weights="generate")[var])  
        dsm[var] = dsm[var]**2
        mec_dict['rmsm'] = np.sqrt(dsm.spatial.average(var,weights="generate")[var])
    
    return mec_dict,dat_mean,dat_std
```

```python
def draw_model_bias_2d(regnam,var,var_dict,frequency,model_list,ref_dict,exp_dict,plot_type,data_path,fig_path,fgw=35,fgh=25,hs=0.5,ws=0.5):
    #generate land mask consistent with MODIS data
    region = define_region(regnam)
    lat_slice = slice(list(region[0])[0],list(region[0])[1])
    lon_slice = slice(list(region[1])[0],list(region[1])[1])
    latreg = list(region[0])
    lonreg = list(region[1])
    extent = [lonreg[0],lonreg[1],latreg[0],latreg[1]]
                       
    if regnam == 'Antarctic': 
        data_projection = ccrs.SouthPolarStereo()
        kwtrans = dict(central_latitude=-90, central_longitude=0.)
        trans=ccrs.PlateCarree()
    elif regnam == 'PolarN':
        data_projection = ccrs.NorthPolarStereo() 
        kwtrans = dict(central_latitude=90, central_longitude=0.)
        trans=ccrs.PlateCarree()
    elif regnam == 'Greenland': 
        data_projection = ccrs.PlateCarree() 
        kwtrans = dict(central_longitude=0.)
        trans = ccrs.PlateCarree(**kwtrans) 
    elif regnam == 'Atlantic': 
        data_projection = ccrs.PlateCarree() #ccrs.Robinson() 
        kwtrans = dict(central_longitude=0.)
        trans = ccrs.PlateCarree(**kwtrans) #ccrs.Stereographic(**kwtrans)
    elif regnam == 'CONUS': 
        lat_0 = 30  # Reference latitude
        lon_0 = -60  # Central longitude
        lat_1 = 20  # First standard parallel
        lat_2 = 40  # Second standard parallel
        data_projection = ccrs.LambertConformal(
            central_longitude=lat_0, 
            central_latitude=lon_0,
            standard_parallels=(lat_1, lat_2)
        )
        kwtrans = dict(central_latitude=39.0, central_longitude=-96.0)
        trans = ccrs.PlateCarree()   
    else:
        data_projection = ccrs.PlateCarree() #ccrs.Robinson() 
        kwtrans = dict(central_longitude=0.)
        trans = ccrs.PlateCarree(**kwtrans) #ccrs.Stereographic(**kwtrans)
                     
    nrows = 1
    ncols = len(model_list)
    
    #create figure  
    frac = 12.0/35.0
    frac = fgh/fgw
    fontz = 14 * fgw*1.0/fgh * frac
    fig, axes = plt.subplots(nrows=nrows,
                             ncols=ncols, 
                             figsize=(fgw,fgh), 
                             subplot_kw={'projection': data_projection})
    
    #cmap = plt.get_cmap('RdBu_r')
    cax = np.empty((nrows,ncols), dtype=plt.Axes)
    contour = np.empty((nrows,ncols), dtype=mcontour.ContourSet)
    
    #------------------------------------------------------------------------------
    # Colormap
    #
    # Generate a colormap with colors from dark blue to dark red with white in the middle.
    #------------------------------------------------------------------------------
    if plot_type == "model_vs_obs":
        color_list4 = ['darkblue', 'lightblue', 'white', 'lightyellow', 'orange',  'firebrick']
        nodes = [0.0, 0.3, 0.5, 0.5, 0.8,  1.0]
        cmap = matcolors.LinearSegmentedColormap.from_list('mybluewhitered', list(zip(nodes, color_list4)))
    else:
        color_list4 = ["darkblue", "blue", "cyan", "green", "yellow", "red"]
        nodes = [0.0, 0.3, 0.5, 0.5, 0.8,  1.0]
        cmap = matcolors.LinearSegmentedColormap.from_list('myrainbow', list(zip(nodes, color_list4)))
        
    vunt = var_dict['unit']
    vmin = min(var_dict['level'])
    vmax = max(var_dict['level'])
    refmodel = var_dict['ref']

    nlev = len(var_dict['level'])
    clevs = var_dict['level']
    

    for j,model in enumerate(model_list):
        
        if plot_type == "model_vs_obs":
            obs_dict = read_obs_var(var,frequency,refmodel,ref_dict[refmodel])
            mod_dict = read_model_var(var,frequency,model,exp_dict,data_path)
            mec_dict,dat_dict,std_dict = derive_hor_metrics_data(var,obs_dict,mod_dict) 
        else: 
            obs_dict = []
            mod_dict = read_model_var(var,frequency,model,exp_dict,data_path)
            mec_dict,dat_dict,std_dict = derive_hor_metrics_data(var,obs_dict,mod_dict) 
            
        if plot_type == "model_vs_obs":
            metrics_str = "RMSE/PCOR: {:0.1f}/{:0.2f}".format(
                mec_dict['rmse'].values,
                mec_dict['pcor'].values
            )
        else:
            metrics_str = "Mean: {:0.1f}".format(
                mec_dict['mean'].values
            )
            
        vplt = dat_dict.copy() 
        print(var,model,metrics_str)
        print(vplt.min().values,vplt.max().values)
        
        k = j
        #------------------------------------------------------------------------------
        # Set data bounds, norm and colorbar labels.
        #------------------------------------------------------------------------------
        cbar_labels = [f'{x:.1f}' for x in clevs]
        cbar_title = '{} ({})'.format(var,vunt)
        norm = matcolors.BoundaryNorm(clevs, cmap.N)
                
        #axes.flat[k].add_feature(cfeature.LAND, facecolor='lightgray')
        axes.flat[k].set_extent(extent, ccrs.PlateCarree())
        axes.flat[k].add_feature(cfeature.COASTLINE, linewidth=0.5, zorder=3)
        axes.flat[k].coastlines()
        axes.flat[k].add_feature(cfeature.LAKES,linewidth=0.5,edgecolor='black',facecolor='None',zorder=4)
            
        kwargs = dict(
                vmin = vmin,
                vmax = vmax,
                levels = clevs,
                colors=cmap, #color_list,
                add_colorbar=False,
                extend='both',
                transform=trans,
                zorder=0)
                
        # Plot filled contours
        contour[0][j] = vplt.plot.contourf(ax=axes.flat[k], **kwargs)
        
        bbox_to_anchor=(0.1, 0, 1, 1)
        labelpad = 8
        xpos_str = 1.0
        ypos_str = 1.0
        cbshk = 0.95
        cbpos = 0.05
        cbort = 'horizontal'
        # Use geocat-viz utility function to format lat/lon tick labels
        gv.add_lat_lon_ticklabels(ax=axes.flat[k])
        gv.set_axes_limits_and_ticks(axes.flat[k],
                                     xlim=(extent[0], extent[1]),
                                     ylim=(extent[2], extent[3]),
                                     xticks=range(extent[0], extent[1]+1, 60),
                                     yticks=range(extent[2], extent[3]+1, 30))
        # Customize ticks and labels
        axes.flat[k].tick_params(labelsize=fontz*1.1, length=8)
        axes.flat[k].set_xlabel("")
        axes.flat[k].set_ylabel("")      
                
        axes.flat[k].set_title('{}'.format(model),loc='left',fontsize=fontz*1.2,pad=10)
        axes.flat[k].set_title('{}'.format(metrics_str),loc='right',fontsize=fontz*1.2,pad=10)
        #axes.flat[k].set_title('{}'.format(model),loc='center',fontsize=fontz*1.2,pad=40)
        #axes.flat[k].text(0.1,1,model,fontsize=fontz*1.2, color='red', fontweight='bold', ha='center', va='top') #transform=plt.gcf().transFigure)
        #axes.flat[k].text(1,1,metrics_str,fontsize=fontz*1.2, color='red', fontweight='bold', ha='center', va='top') #transform=plt.gcf().transFigure)
        #axes.flat[k].text(xpos_str,ypos_str,metrics_str,color='red',fontsize=fontz,fontweight='bold',ha='center',va='center',transform=axes.flat[k].transAxes)
        cbar = plt.colorbar(contour[0][j],
                            cmap=cmap,
                            norm=norm,
                            extend='both',
                            orientation= cbort,
                            shrink = cbshk,
                            pad = cbpos)
        cbar.set_label(cbar_title,rotation=0,labelpad=labelpad,weight='normal',fontsize=fontz*1.2)
        cbar.ax.tick_params(labelsize=fontz*1.1)
        cbar.set_ticklabels(cbar_labels, weight='normal')
        if len(clevs) < 12: 
            cbar.solids.set(edgecolor='black', linewidth=0.2)
            cbar.outline.set(edgecolor='black', linewidth=0.2)
            

    #plt.tight_layout()
    plt.subplots_adjust(hspace=hs)
    plt.subplots_adjust(wspace=ws)
    plt.show()
    
    #plt.show()
    if plot_type == "model_vs_obs":
        fig.savefig('fig_2d_bias_{}_{}_{}_{}_{}.pdf'.format(plot_type,refmodel,var,regnam,exp_dict[model]['period']))
    else:
        fig.savefig('fig_2d_state_{}_{}_{}_{}_{}.pdf'.format(plot_type,refmodel,var,regnam,exp_dict[model]['period']))
    # Close the current figure
    plt.close()
    
    return 
```
