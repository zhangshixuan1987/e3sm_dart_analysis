```python
import os
import xarray as xr
import matplotlib.pyplot as plt

class SEBMultiExperimentPlotter:
    def __init__(self, base_dir, experiments, ensemble, region='global', frequency='monthly'):
        """
        Parameters:
        - base_dir: Path to processed SEB outputs.
        - experiments: List of experiment names (e.g., ['CTRLEN10', 'DARTEN20']).
        - ensemble: Common ensemble ID (e.g., 'EN01').
        - region: 'global', 'land', or 'ocean'.
        - frequency: 'daily' or 'monthly'.
        """
        self.data = {}
        self.region = region
        self.ensemble = ensemble
        self.frequency = frequency

        for exp in experiments:
            file_path = os.path.join(base_dir, exp, f"seb_{frequency}_{ensemble}_{region}.nc")
            if os.path.exists(file_path):
                self.data[exp] = xr.open_dataset(file_path)
            else:
                print(f"[WARN] File not found for {exp}: {file_path}")

    def plot_timeseries_comparison(self, variable='Rn'):
        """
        Compare time series of a variable across experiments.
        """
        plt.figure(figsize=(12, 5))
        for exp, ds in self.data.items():
            ds_mean = ds[variable].mean(dim=['lat', 'lon'])
            ds_mean.plot(label=exp)
        plt.title(f"{self.ensemble} | {self.region.upper()} | {variable} Time Series")
        plt.ylabel("W/m²")
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

    def plot_monthly_climatology_comparison(self, variable='Rn'):
        """
        Plot monthly climatology (mean over time) for each experiment.
        """
        plt.figure(figsize=(10, 5))
        for exp, ds in self.data.items():
            ds_mean = ds[variable].mean(dim=['lat', 'lon'])
            clim = ds_mean.groupby("time.month").mean("time")
            plt.plot(clim['month'], clim, label=exp)
        plt.title(f"{self.ensemble} | {self.region.upper()} | Monthly Climatology of {variable}")
        plt.xlabel("Month")
        plt.ylabel("W/m²")
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

    def plot_bowen_ratio_climatology(self):
        """
        Plot monthly climatology of Bowen Ratio across experiments.
        """
        if not any("BowenRatio" in ds for ds in self.data.values()):
            print("[WARN] BowenRatio not available in all datasets.")
            return

        plt.figure(figsize=(10, 5))
        for exp, ds in self.data.items():
            if "BowenRatio" in ds:
                br = ds["BowenRatio"].mean(dim=["lat", "lon"])
                br_clim = br.groupby("time.month").mean("time")
                plt.plot(br_clim['month'], br_clim, label=exp)
        plt.title(f"{self.ensemble} | {self.region.upper()} | Monthly Climatology of Bowen Ratio")
        plt.xlabel("Month")
        plt.ylabel("SH / LE")
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()
```
